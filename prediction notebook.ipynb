```python
#change iphone format to jpg
from PIL import Image
import pillow_heif
import cv2
import os

# Register HEIF opener so PIL can read .heic directly
pillow_heif.register_heif_opener()

# Input & output folders
input_dir = '/Users/xb/Desktop/Uni Notes/Y3S1/SC2079/texture images'
output_dir = os.path.join(input_dir, "converted")
os.makedirs(output_dir, exist_ok=True)

# Loop through files
for file in os.listdir(input_dir):
    if file.lower().endswith(".heic"):
        img_path = os.path.join(input_dir, file)

        # Open HEIC image directly with Pillow
        img = Image.open(img_path)

        # Save as JPEG
        out_path = os.path.join(output_dir, file.rsplit(".", 1)[0] + ".jpg")
        img.save(out_path, "JPEG", quality=95)

        print(f"Converted {file} → {out_path}")
```

```python
#change to grey scale(not necessary)
import cv2
import os

# Input & output paths
input_dir = '/Users/xb/Desktop/Uni Notes/Y3S1/SC2079/texture images/converted‘
output_dir = '/Users/xb/Desktop/Uni Notes/Y3S1/SC2079/self taken image/allcards/BW'
os.makedirs(output_dir, exist_ok=True)

# Loop through images
for file in os.listdir(input_dir):
    if file.lower().endswith((".jpg", ".jpeg", ".png")):
        img_path = os.path.join(input_dir, file)

        # Read and convert to grayscale
        img = cv2.imread(img_path)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Convert single channel → 3-channel (YOLO expects 3 channels)
        gray_3ch = cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR)

        # Save processed image (keep same filename)
        save_path = os.path.join(output_dir, file)
        cv2.imwrite(save_path, gray_3ch)

        print(f"Processed {file} → saved as {save_path}")
```

```python
# Load your trained model
model = YOLO('/Users/xb/Desktop/Uni Notes/Y3S1/SC2079/task 1/bestL160epoch.pt') #model path

# Run prediction on an image
results = model.predict(
    source= "/Users/xb/Downloads/pic", #input file path
    save=True,
    project='results', #output file in the notebook directory 
    name='prediction',
    conf= 0.3,
)
```

```python
#print result if output image overlap

class_names = [
    'A','B','Bullseye','C','D','E','F','G','H','S','T','U','V','W','X','Y','Z',
    'circle','down','eight','five','four','left','nine','one','right',
    'seven','six','three','two','up'
]

for r in results:
    image_name = r.path.split("/")[-1]   # just the file name
    confs = r.boxes.conf.cpu().numpy()   # confidence scores
    clss  = r.boxes.cls.cpu().numpy().astype(int)   # class indices

    print(f"\nImage: {image_name}")
    for conf, cls_idx in zip(confs, clss):
        class_name = class_names[cls_idx]
        print(f"  Class: {class_name} | Conf: {conf:.2f}")
```

```python
#no use
import cv2
import numpy as np

def resize_one_image(input_path, output_path):
    img = cv2.imread(input_path)
    h, w = img.shape[:2]
    
    # Compute resize scale and new size
    scale = 640 / max(h, w)
    new_w, new_h = int(w * scale), int(h * scale)
    resized = cv2.resize(img, (new_w, new_h), interpolation=cv2.INTER_LINEAR)
    
    # Add padding to make it 640x640
    top = (640 - new_h) // 2
    bottom = 640 - new_h - top
    left = (640 - new_w) // 2
    right = 640 - new_w - left
    padded = cv2.copyMakeBorder(resized, top, bottom, left, right, cv2.BORDER_CONSTANT, value=(114,114,114))
    
    # Save
    cv2.imwrite(output_path, padded)

# Example usage
resize_one_image('/Users/xb/Downloads/Photo on 29-8-25 at 9.20 AM.jpg', '/Users/xb/Downloads/Photo on 29-8-25 at 9.20 AM.jpg')
```
